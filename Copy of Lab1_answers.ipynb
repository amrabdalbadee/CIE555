####(1) Importing required libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import io
#from google.colab import files
#use sklearn
from sklearn.datasets import load_iris
```

####(2) Loading data 

```python
#read data from csv file
#uploaded = files.upload()
#iris = pd.read_csv('Iris.csv')
iris_data = load_iris()
iris = pd.DataFrame(data= np.c_[iris_data['data'], iris_data['target']],
                     columns= iris_data['feature_names'] + ['target'])
iris.head()
#iris
```

####(3) Helper functions

```python
#Convert array to one-hot encoding


# Write a function that takes an array of numeric classes and turns it into
# one-hot encoded array meaning that if we have [0,1,2], 
# it becomes [[1,0,0],[0,1,0],[0,0,1]]
def to_one_hot(Y):
    n_col = np.amax(Y) + 1
    binarized = np.zeros((len(Y), n_col))
    for i in range(len(Y)):
        binarized[i, Y[i]] = 1.
    return binarized
##############################
  

# Write a function that takes an array and performs the sigmoid function on all 
# its elements
# make it take two arguments, the firs one is the matrix and the second one
# is a boolean input "derivative" which if true, it computer the derivative of
# the sigmoid function. Make the default to be false
def sigmoid(x, deriv = False):
  sig = 1/(1+np.exp(-x))
  if(deriv==True):
    sig = sig*(1-sig)
  return sig
##############################
  
  
##Normalize array
def normalize(X, axis=1, order=2):
    l2 = np.atleast_1d(np.linalg.norm(X, order, axis))
    l2[l2 == 0] = 1
    return X / np.expand_dims(l2, -1)
```

####(4) Preparing data (Preprocessing)

```python
x = iris[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']]
x1 = normalize(x.values)
```

```python
y = iris['target']
#Convert the DataFrame to NumPy array
y = y.values
#Cast the values of the target to int type
y = y.astype(int)
#Convert to one-hot encoded array size=(no_instances,no_attributes)
y = to_one_hot(y)
```

####(5) Training

```python
#Split data to training and validation data
X_train, X_test, y_train, y_test = train_test_split(x1, y, test_size=0.33)

###Main network with 2 layers###

#number of iterations
iter = 50000 


#Initialize the weights of the first and second layers: Determine the size of the 2D array
#Weights
w0 = np.random.normal(size = (4, 5))
w1 = np.random.normal(size = (5, 3))
##############################


#Initialize the bias of the first layer: Determine the size of the 1D array
w0_dims = w0.shape
(r_w0,c_w0) = w0_dims
b0 = np.zeros((1,c_w0))


#learning rate
sigma = 0.1

#Errors array for every iteration
errors = []

#Train
for i in range(iter):


    # the input to this function is the input data and the output is the label
    #Feed forward
    
    a0 = X_train
    a1 = sigmoid(np.dot(a0, w0))
    a2 = sigmoid(np.dot(a1, w1))
    ##############################

    #Back propagation using gradient descent
    a2_error = y_train - a2
    a2_delta = a2_error * sigmoid(a2, deriv=True)
    
    a1_error = a2_delta.dot(w1.T)
    a1_delta = a1_error * sigmoid(a1, deriv=True)
    
    w1 += a1.T.dot(a2_delta) * sigma
    w0 += a0.T.dot(a1_delta) * sigma
    
    error = np.mean(np.abs(a2_error))
    errors.append(error)
    accuracy = (1 - error) * 100

#Plot the accuracy graph
plt.plot(errors)
plt.xlabel('Training')
plt.ylabel('Error')
plt.show()
        
print("Training Accuracy " + str(round(accuracy,2)) + "%")
```
